```python
# 02_eda.ipynb
# Exploratory Data Analysis for Education Performance Analytics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load cleaned dataset
df = pd.read_csv(r"C:\Projects\education-performance-analytics\data\student_performance_socioeconomic_cleaned.csv")

# Quick inspection
print(df.shape)
print(df.info())
print(df.describe())

# Histogram of all features
df.hist(figsize=(14, 10))
plt.tight_layout()
plt.savefig(r"C:\Projects\education-performance-analytics\visuals\histogram_all_features.png", dpi=300)
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numeric Features")
plt.savefig(r"C:\Projects\education-performance-analytics\visuals\heatmap_correlations.png", dpi=300)
plt.show()

# Pairplot of academic relationships
sns.pairplot(df, vars=["Attendance", "Semester_Average_Grade", "Grade_Average"], hue="Gender", diag_kind="kde")
plt.suptitle("Pairwise Academic Relationships by Gender", y=1.02)
plt.savefig(r"C:\Projects\education-performance-analytics\visuals\pairplot_academic_relationships.png", dpi=300)
plt.show()
```
